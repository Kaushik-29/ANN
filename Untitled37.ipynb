```python
import numpy as np
import pandas as pd
import re
from time import time
import pickle

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier

# -----------------------------
# 1. Load dataset
# -----------------------------
df = pd.read_csv('/content/final_dataset_hinglish.csv')  # Update path
# Columns available based on previous successful runs: headline, label

# Convert labels to 0/1 if needed
df['label'] = df['label'].replace(-1, 1)

# -----------------------------
# 2. Clean tweet text
# -----------------------------
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'@[\w]*','', text)            # remove mentions
    text = re.sub(r'[^a-zA-Z]', ' ', text)      # remove non-letters
    text = ' '.join([w for w in text.split() if len(w)>2])  # remove short words
    return text

df['tidy_tweet'] = df['headline'].apply(clean_text)

# -----------------------------
# 3. Feature extraction
# -----------------------------
# 3a. Text -> CountVectorizer
vectorizer = CountVectorizer(stop_words='english', lowercase=True)
X = vectorizer.fit_transform(df['tidy_tweet']).toarray()
y = df['label'].values


# -----------------------------
# 4. Train/test split
# -----------------------------
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

# -----------------------------
# 5. ML Pipeline
# -----------------------------
def pipeline(X_train, y_train, X_test, y_test):
    learner_list = [LinearSVC(), RandomForestClassifier(), LogisticRegression(max_iter=1000), KNeighborsClassifier()]

    final_results = []

    for learner in learner_list:
        print(f"Training {learner.__class__.__name__}...")
        results = {}
        results['Algorithm'] = learner.__class__.__name__

        start = time()
        learner.fit(X_train, y_train)
        end = time()
        results['Training Time'] = end - start

        # Save model
        pickle.dump(learner, open(f"{learner.__class__.__name__}.pkl", 'wb'))

        # Predictions
        start = time()
        y_pred_test = learner.predict(X_test)
        y_pred_train = learner.predict(X_train)
        end = time()
        results['Prediction Time'] = end - start

        # Metrics
        results['Accuracy: Test'] = accuracy_score(y_test, y_pred_test)
        results['Accuracy: Train'] = accuracy_score(y_train, y_pred_train)
        results['Precision: Test'] = precision_score(y_test, y_pred_test)
        results['Precision: Train'] = precision_score(y_train, y_pred_train)
        results['Recall: Test'] = recall_score(y_test, y_pred_test)
        results['Recall: Train'] = recall_score(y_train, y_pred_train)
        results['F1 Score: Test'] = f1_score(y_test, y_pred_test)
        results['F1 Score: Train'] = f1_score(y_train, y_pred_train)


        final_results.append(results.copy())
        print(f"{learner.__class__.__name__} Done!\n")


    return final_results

final_results = pipeline(X_train, y_train, X_test, y_test)

# -----------------------------
# 6. Final results table
# -----------------------------
results_df = pd.DataFrame(final_results)
results_df = results_df.sort_values(by='F1 Score: Test', ascending=False).reset_index(drop=True)
print(results_df)

# -----------------------------
# 7. Predict new sentence
# -----------------------------
def predict_new_tweet(tweet_text, vectorizer, model_filename):
    # Clean tweet
    clean_tweet = clean_text(tweet_text)

    # Text -> vector
    X_new = vectorizer.transform([clean_tweet]).toarray()

    # Load model
    model = pickle.load(open(model_filename, 'rb'))

    pred = model.predict(X_new)[0]
    return "Bully" if pred==1 else "Non-Bully"

# Example usage
example = "You are not idiot, don't ever feel like that"
print("Prediction:", predict_new_tweet(example, vectorizer=vectorizer, model_filename="LinearSVC.pkl"))
```

```python
import pandas as pd
import torch
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split
from transformers import DistilBertTokenizerFast, DistilBertForSequenceClassification, Trainer, TrainingArguments
from sklearn.metrics import accuracy_score, precision_recall_fscore_support
from transformers.integrations import TrainerCallback
from tqdm.notebook import tqdm

# -------------------------------
# 1. Load dataset
# -------------------------------
df = pd.read_csv("/content/cleaned_dataframe.csv")
print("Columns available:", df.columns.tolist())

df['Bully_Label'] = df['Bully_Label'].map({'Bully': 1, 'Non_bully': 0})
df['cleaned_tweet'] = df['cleaned_tweet'].fillna('')
df.dropna(subset=['Bully_Label'], inplace=True)

# Train/test split
train_texts, val_texts, train_labels, val_labels = train_test_split(
    df['cleaned_tweet'].tolist(),
    df['Bully_Label'].tolist(),
    test_size=0.2,
    random_state=42,
    stratify=df['Bully_Label']
)

# -------------------------------
# 2. Tokenizer
# -------------------------------
tokenizer = DistilBertTokenizerFast.from_pretrained("distilbert-base-uncased")

train_encodings = tokenizer(train_texts, truncation=True, padding=True, max_length=128)
val_encodings = tokenizer(val_texts, truncation=True, padding=True, max_length=128)

# -------------------------------
# 3. Torch Dataset
# -------------------------------
class BullyDataset(Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels
    def __len__(self):
        return len(self.labels)
    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        # Convert labels to Long type
        item['labels'] = torch.tensor(self.labels[idx], dtype=torch.long)
        return item

train_dataset = BullyDataset(train_encodings, train_labels)
val_dataset = BullyDataset(val_encodings, val_labels)

# -------------------------------
# 4. Model
# -------------------------------
# Ensure num_labels is 2 for binary classification
model = DistilBertForSequenceClassification.from_pretrained("distilbert-base-uncased", num_labels=2)

# -------------------------------
# 5. Metrics
# -------------------------------
def compute_metrics(pred):
    labels = pred.label_ids
    preds = pred.predictions.argmax(-1)
    acc = accuracy_score(labels, preds)
    precision, recall, f1, _ = precision_recall_fscore_support(labels, preds, average='binary')
    return {"accuracy": acc, "precision": precision, "recall": recall, "f1": f1}

# -------------------------------
# 6. Training
# -------------------------------
training_args = TrainingArguments(
    output_dir="./results",
    eval_strategy="epoch",
    save_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=3,
    weight_decay=0.01,
    logging_dir="./logs",
    logging_steps=10,
    load_best_model_at_end=True,
    metric_for_best_model="f1"
)

# Add a simple progress bar callback
class TqdmCallback(TrainerCallback):
    def on_epoch_begin(self, args, state, control, **kwargs):
        if state.is_world_process_zero:
            self.pbar = tqdm(total=state.max_steps, desc=f"Epoch {int(state.epoch) + 1}")

    def on_step_end(self, args, state, control, **kwargs):
        if state.is_world_process_zero:
            self.pbar.update(1)
            if state.global_step % args.logging_steps == 0:
                self.pbar.write(f"Step {state.global_step}/{state.max_steps} - Loss: {state.log_history[-1]['loss']:.4f}")


    def on_epoch_end(self, args, state, control, **kwargs):
      if state.is_world_process_zero:
        self.pbar.close()


trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics,
    callbacks=[TqdmCallback] # Add the callback here
)

trainer.train()

# Save the model and tokenizer after training
trainer.save_model("./fine-tuned-model")
tokenizer.save_pretrained("./fine-tuned-model")


# -------------------------------
# 7. Inference Function
# -------------------------------
def predict_text(text):
    # Load the fine-tuned model and tokenizer
    loaded_tokenizer = DistilBertTokenizerFast.from_pretrained("./fine-tuned-model")
    loaded_model = DistilBertForSequenceClassification.from_pretrained("./fine-tuned-model")

    inputs = loaded_tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=128)
    with torch.no_grad():
        outputs = loaded_model(**inputs)
        logits = outputs.logits
        pred = torch.argmax(logits, dim=-1).item()
        prob = torch.softmax(logits, dim=-1).tolist()[0]
    print(f"\nTweet: {text}")
    print(f"Prediction: {'Bully' if pred==1 else 'Non-bully'}")
    print(f"Probabilities: Non-bully={prob[0]:.3f}, Bully={prob[1]:.3f}")
    return pred

# -------------------------------
# 8. Example Tests
# -------------------------------
predict_text("You are so stupid!")
predict_text("do not feel bad, you are not idiot")
predict_text("Wow, amazing genius move!")
```

```python
import pandas as pd
import torch
from torch.utils.data import Dataset
from sklearn.model_selection import train_test_split
from transformers import DistilBertTokenizerFast, DistilBertForSequenceClassification, Trainer, TrainingArguments
from sklearn.metrics import accuracy_score, precision_recall_fscore_support # Import metrics

# 1. Load dataset
df = pd.read_csv("/content/cleaned_dataframe.csv")  # Replace with your dataset file
print(df.head())

# Handle missing values in 'Bully_Label' before mapping and creating texts list
df.dropna(subset=['Bully_Label'], inplace=True)

# Use 'cleaned_tweet' as input text and 'Bully_Label' as target
texts = df["cleaned_tweet"].astype(str).tolist()

# Map 'Bully' to 1 and 'Non_bully' to 0 before converting to int
labels = df["Bully_Label"].map({'Bully': 1, 'Non_bully': 0}).astype(int).tolist()

# Train-test split
train_texts, val_texts, train_labels, val_labels = train_test_split(
    texts, labels, test_size=0.2, random_state=42
)

# 2. Load tokenizer
tokenizer = DistilBertTokenizerFast.from_pretrained("distilbert-base-uncased")

train_encodings = tokenizer(train_texts, truncation=True, padding=True, max_length=128)
val_encodings = tokenizer(val_texts, truncation=True, padding=True, max_length=128)

# 3. Torch Dataset class
class BullyDataset(Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item["labels"] = torch.tensor(self.labels[idx])
        return item

train_dataset = BullyDataset(train_encodings, train_labels)
val_dataset = BullyDataset(val_encodings, val_labels)

# 4. Load model
model = DistilBertForSequenceClassification.from_pretrained("distilbert-base-uncased", num_labels=2)

# 5. Metrics
def compute_metrics(pred):
    labels = pred.label_ids
    preds = pred.predictions.argmax(-1)
    acc = accuracy_score(labels, preds)
    precision, recall, f1, _ = precision_recall_fscore_support(labels, preds, average='binary')
    return {"accuracy": acc, "precision": precision, "recall": recall, "f1": f1}


# 6. Training arguments
training_args = TrainingArguments(
    output_dir="./results",
    eval_strategy="epoch", # Corrected argument name
    save_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=3,
    weight_decay=0.01,
    logging_dir="./logs",
    logging_steps=10,
    load_best_model_at_end=True,
    metric_for_best_model="f1" # Added metric for best model
)

# 7. Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics # Added compute_metrics
)

# 8. Train model
trainer.train()

# 9. Save model
trainer.save_model("./bully_detection_model")
tokenizer.save_pretrained("./bully_detection_model")

# 10. Inference example
def predict(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=128)
    outputs = model(**inputs)
    preds = torch.argmax(outputs.logits, dim=1)
    return "Bully" if preds.item() == 1 else "Non-bully"

print(predict("You are so stupid!"))
```
